```python
import numpy as np
import pandas as pd
from time import time
import tensorflow as tf
from tensorflow.keras.models import * 
from tensorflow.keras.layers import * 
from tensorflow.keras.utils  import * 
from datetime import datetime
%matplotlib inline
import matplotlib.pyplot as plt

# Data preprocessing

def preprocess(path ='../ml-1m/', 
               file = 'ratings.dat', 
               mode = 'test',               # validation / test
               core = 1,                    # min number of item actions
               sequence_length = 5,         # length of input sequence for training and prediction
               n_test_neg = 100,            # number of negative samples per positive for test
               n_pos = 5,                   # number of positive samples for training = n_pos * len(user_train)
               verbose = 0,                 # > 0 - more diagnostics 
               users = 0):                  # how many users to process, 0 - all users

    ratings = pd.read_table(path + file, sep='::', engine='python', 
                    names=['userid', 'itemid', 'rating', 'timestamp']).drop(columns=['rating']).astype('int32')
    
    n_users = ratings.userid.nunique()
    n_items = ratings.itemid.nunique()

    print('users   in dataset\t', n_users)
    print('items   in dataset\t', n_items)
    print('actions in dataset\t', ratings.shape[0])

    t_preprocess = time()

    if core > 0: # remove items with less than core actions
        ratings = ratings.groupby(['itemid'], group_keys=False).filter(lambda x: len(x) >= core)

    ratings['itemid'], itemid_uniques = pd.factorize(ratings['itemid'])
    ratings['itemid'] += 1
    ratings = ratings.astype('int32')

    all_items_set = set(ratings['itemid'])

    gu = ratings.groupby(['userid'], group_keys=False) 

    n_items = ratings.itemid.nunique()
    print('items with actions >=', core, 'after filtering\t', n_items)
    print(45*'=')

    if verbose > 0:
        display(ratings)
        gt = ratings.groupby(['timestamp'], group_keys=False)
        print('number of equal timestamps\n', gt.size().sort_values())
        print('\nratings for users\n', gu.size())
        display(gu.get_group(1).sort_values('timestamp', ignore_index=True))


    print('\nPreparation of Negative and Positive Samples for Training and Testing\n')

    users_train_list = []
    users_test_list  = []
    users_train_target_list = []
    users_test_target_list  = []

    u = 0
    
    if users == 0:
        users = len(gu)
        
    for user_index in range(1, users + 1):

        user = gu.get_group(user_index).sort_values('timestamp', ignore_index=True).drop(
            columns=['timestamp']).to_numpy(dtype='int32')

        if mode == 'validation':
            user = user[ :-1]                    # remove test row, keep validation data only

        user_seen = set(user[ :-1, 1])
        user_unseen = list(all_items_set - user_seen)

        len_unseen = len(user_unseen)

        target = user[sequence_length : , 1:2]   # target column (shifted items)
        items  = user[ : , 1:2]                  # items column

        user_length = len(user) - sequence_length
        user = user[ : user_length, 0:2]

        for i in range(1, sequence_length):
            user = np.hstack((user, items[i : user_length + i]))

        user = np.hstack((user, target))

        user_train = user[ :-1]                  # train data without last row
        user_test  = user[-1]                    # last row with true value for test

        if len(user_train) > len_unseen:
            replace = True
            l = 2
        else:
            replace = False
            l = n_pos

        user_train_len = len(user_train)

        user_train__ = np.zeros_like(user_train, shape=(l * user_train_len + len_unseen, user_train.shape[1]))
        user_train_fill = 0

        while user_train_fill+user_train_len <= user_train__.shape[0]:
            user_train__[user_train_fill : user_train_fill+user_train_len] = user_train
            user_train_fill += user_train_len

        if user_train_fill < user_train__.shape[0]:
            user_train__[user_train_fill : ] = user_train[-user_train__.shape[0]+user_train_fill : ]    

        user_train = user_train__

        user_train[l*user_train_len:,-1]=np.random.choice(user_unseen,len_unseen,replace=replace).astype('int32')

        train_target = np.zeros((user_train.shape[0]), dtype='int32')
        train_target[ : l * user_train_len].fill(1)

        user_unseen.remove(user_test[-1])        # test row with true value should not be in test negatives

        user_test = np.tile(user_test, (n_test_neg + 1, 1))
        user_test[1:, -1] = np.random.choice(user_unseen, user_test.shape[0]-1, replace=False).astype('int32')

        test_target = np.zeros((user_test.shape[0]), dtype='int32')
        test_target[0] = 1 # positive


        users_train_list.append(user_train)
        users_test_list.append(user_test)

        users_train_target_list.append(train_target)
        users_test_target_list.append(test_target)

        u += 1

        if verbose > 0:

            if u == 1:
                print('user_train\n', user_train)
                print('\ntrain_target\n', train_target)
                print('\nuser_test\n', user_test)
                print('\ntest_target\n', test_target)            

        if u % 1000 == 0:
            print('Train & Test data. Users processed:', u, '\ttime =', round(time()-t_preprocess, 1))

    print('Train & Test data prepared. Users processed:', u, '\ttime =', round(time()-t_preprocess, 1), '\n')

    users_train = np.concatenate((users_train_list))
    users_test  = np.concatenate((users_test_list))

    train_targets = np.concatenate((users_train_target_list))
    test_targets  = np.concatenate((users_test_target_list))

    train_positives = np.sum(train_targets)
    train_negatives = train_targets.shape[0] - train_positives

    print('train_positives\t', train_positives)
    print('train_negatives\t', train_negatives)
    print('train_negatives/train_positives =', round((train_negatives/train_positives), 1))

    print('\nusers_train.shape\t', users_train.shape)
    print('train_targets.shape\t', train_targets.shape)

    print('\nusers_test.shape\t', users_test.shape)
    print('test_targets.shape\t', test_targets.shape)

    return n_users, n_items, users_train,  train_targets, users_test, test_targets, users_test_list


def hit_ndcg_test(users_test, n_users_test, hit_size=10):
    
    hit  = 0.
    ndcg = 0.
    
    if (n_users_test <= 0) or (n_users_test > len(users_test)):
        n_users_test = len(users_test)
        
    t1 = time()
    u = 0
    for user_test in users_test_list[:n_users_test]:

        predicted_weights = model.predict([user_test[:, 0], user_test[:, 1:]], verbose=0)
        recommended = np.argsort((-predicted_weights).flat)[:hit_size]

        ii = np.where(recommended == 0)

        if len(ii[0]) > 0:
            true_item_index = ii[0][0] 
            ndcg += 1./np.log2(true_item_index+2)    # true_item_index = 0...hit_size-1
            hit  += 1
        
        u += 1
    
    return hit/u, ndcg/u


# Model preprocessing, training and testing

gpus = tf.config.experimental.list_physical_devices('GPU')
print(gpus, '\n', 65*'=')
tf.config.experimental.set_memory_growth(gpus[0], True)
print('Tensorflow version', tf.__version__, '\n')


run = 3                 # how many times run preprocessing, training and testing

sequence_length = 5
embedding_size = 40
dropout   = 0.5
epochs    = 5
batchsize = 128
layersize = 2240
    
t_all_run = time()


core = 11

hit_array  = np.zeros((core, run, epochs), dtype=float)
ndcg_array = np.zeros((core, run, epochs), dtype=float)


for c in range(core):
    print('\n', c, '- core')
    
    for r in range(run):
        print(72*'=')

        n_users, n_items, users_train, train_targets, users_test, test_targets, users_test_list = preprocess(
                    core = c,                # min number of item actions
                    sequence_length = sequence_length, # length of input sequence for training and prediction
                    n_test_neg = 100,        # number of negative samples per positive for test
                    n_pos = 5,               # number of positive samples for training = n_pos * len(user_train)
                    users = 0)               # how many users to process, 0 - all users

        t_run  = time()

        users_input = Input(shape=(1,), name='users_input')
        items_input = Input(shape=(sequence_length+1,), name='items_input')

        users_embedding = Embedding(n_users+1, embedding_size, input_length=1, name='users_embedding')(users_input)
        items_embedding = Embedding(n_items+1, embedding_size, input_length=sequence_length+1, 
                                    name='items_embedding')(items_input)

        users_and_items = concatenate([users_embedding, items_embedding], axis=1, name='users_and_items')
        users_and_items_flat = Flatten(name='users_and_items_flat')(users_and_items)

        nn = LayerNormalization()(users_and_items_flat)
        nn = Dense(layersize, activation='relu')(nn)
        nn = LayerNormalization()(nn)
        nn = Dropout(dropout)(nn)

        nn0 = Dense(layersize//2, activation='relu')(nn)
        nn0 = LayerNormalization()(nn0)
        nn0 = Dropout(dropout//2)(nn0)

        nn1 = Dense(layersize//4, activation='relu')(nn)
        nn1 = LayerNormalization()(nn1)
        nn1 = Dropout(dropout//4)(nn1)

        nn2 = Dense(layersize//8, activation='relu')(nn)
        nn2 = LayerNormalization()(nn2)
        nn2 = Dropout(dropout//8)(nn2)

        nn3 = Dense(layersize//14, activation='relu')(nn)
        nn3 = LayerNormalization()(nn3)

        nn4 = Dense(layersize//28, activation='relu')(nn)
        nn4 = LayerNormalization()(nn4)

        nn5 = Dense(layersize//56, activation='relu')(nn)
        nn5 = LayerNormalization()(nn5)

        nn6 = Dense(layersize//112, activation='relu')(nn)
        nn6 = LayerNormalization()(nn6)

        nn7 = Dense(layersize//224, activation='relu')(nn)
        nn7 = LayerNormalization()(nn7)

        concat = concatenate([nn0, nn1, nn2, nn3, nn4, nn5, nn6, nn7], axis=1, name='concatenated')
        concat = LayerNormalization()(concat)

        concat = Dense(layersize//2, activation='relu')(concat) 

        prediction = Dense(1, activation='sigmoid', name='prediction')(concat)

        model = Model(inputs=[users_input, items_input], outputs=prediction)

        model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
        
        if (c == 0) and (r == 0):
            print(model.summary())

        print(r+1, '\t') 

        hit_max  = 0.
        ndcg_max = 0.

        try:
            for epoch in range(1, epochs+1):
                print('Epoch ', epoch)

                training = model.fit([users_train[:, 0], users_train[:, 1:]], train_targets, epochs=1, 
                                     batch_size=batchsize, verbose=2, shuffle=True)

                hit, ndcg = hit_ndcg_test(users_test, len(users_test_list))

                print('\t    Hit@10 =', round(hit, 4), '\tNDCG@10 =', round(ndcg, 4))

                hit_max  = max(hit_max, hit)
                ndcg_max = max(ndcg_max, ndcg)

                hit_array[c, r, epoch-1]  = hit
                ndcg_array[c, r, epoch-1] = ndcg

        except KeyboardInterrupt: 
            pass

    print(72*'=', '\n')
    print('Train + test time =', round(time()-t_all_run))

    for i in range(epochs):
        print('Epoch', i+1, ': \tHit@10 =', round(np.mean(hit_array[c, :, i]), 4), 
                            '  \tNDCG@10 =', round(np.mean(ndcg_array[c, :, i]), 4))
    print(72*'=', '\n')
    
print(72*'=', '\n')

hits_ = []
ndcg_ = []

for i in range(core):
    hits_.append(np.mean(hit_array[i, :, epochs-1]))
    ndcg_.append(np.mean(ndcg_array[i, :, epochs-1]))
    
    print(i, '- core: \tHit@10 =', round(np.mean(hit_array[i, :, epochs-1]), 4), 
                        '  \tNDCG@10 =', round(np.mean(ndcg_array[i, :, epochs-1]), 4))
print(72*'=', '\n')

plt.plot(hits_, label='Hit@10')
plt.plot(ndcg_, label='NDCG@10')
plt.legend()
```
